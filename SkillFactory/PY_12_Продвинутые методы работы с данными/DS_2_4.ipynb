## Закрепление знаний
### Для решения задач нам понадобится выделить из признака title год выпуска фильма. Для этого напишем функцию get_year_release(arg).

```python
import pandas as pd
import os
rating = pd.read_csv('data/movies_data/ratings_movies.csv', sep=',')
rating.head()
```

```python
ratings = rating.copy() # СОЗДАНИЕ КОПИИ ТАБЛИЦЫ ! ! !
ratings.head()
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
# Создайте в таблице новый признак year_release, 
# который соответствует году выпуска фильма.
# У скольких фильмов не указан год их выпуска?
ratings['year_release'] = ratings['title'].apply(get_year_release)
ratings.info()
# ответ - у 18-ти
```

```python
# Какой фильм, выпущенный в 1999 году,
# получил наименьшую среднюю оценку зрителей?
# В качестве ответа запишите название этого фильма без указания года его выпуска.
mask = ratings['year_release'] == 1999
ratings[mask].groupby('title')['rating'].mean().sort_values()
```

```python
# Какое сочетание жанров фильмов (genres), выпущенных в 2010 году, 
# получило наименьшую среднюю оценку (rating)?
# Запишите сочетание так же, как оно указано в таблице (через разделитель |, без пробелов).
mask = ratings['year_release'] == 2010
ratings[mask].groupby('genres')['rating'].mean().sort_values()
```

```python
# Какой пользователь (userId) посмотрел наибольшее количество 
# различных (уникальных) жанров (genres) фильмов? 
# В качестве ответа запишите идентификатор этого пользователя.
ratings.groupby('userId')['genres'].nunique().sort_values(ascending=False)
```

```python
# Найдите пользователя, который выставил наименьшее количество оценок, но его средняя оценка
# фильмам наибольшая. В качестве ответа укажите идентификатор этого пользователя.
# Чтобы рассчитать несколько параметров для каждого пользователя (количество оценок и среднюю оценку),
# можно воспользоваться методом agg() на сгруппированных данных.
ratings.groupby('userId')['rating'].agg(
    ['count', 'mean']
).sort_values(['count', 'mean'], ascending=[True, False])
```

```python
# Найдите сочетание жанров (genres) в 2018 году, которое имеет наибольший средний рейтинг 
# (среднее по столбцу rating), и при этом число выставленных ему оценок (количество значений в
# столбце rating) больше 10. Запишите сочетание так же, как оно указано в таблице
# (через разделитель |, без пробелов).
mask = ratings['year_release'] == 2018
grouped = ratings[mask].groupby('genres')['rating'].agg(
    ['mean', 'count']
)
grouped[grouped['count']>10].sort_values(
    by='mean',
    ascending=False
)
```

```python
# Добавьте в таблицу новый признак year_rating — год выставления оценки.
# Создайте сводную таблицу, которая иллюстрирует зависимость среднего рейтинга 
# фильма от года выставления оценки и жанра. Выберите верные варианты ответа,
# исходя из построенной таблицы:
ratings['date'] = pd.to_datetime(ratings['date'])
ratings['year_rating'] = ratings['date'].dt.year
pivot = ratings.pivot_table(
    index='year_rating',
    columns='genres',
    values='rating',
    aggfunc='mean'
)
display(pivot)
# варианты ответов:
# A За весь период (с 1996 по 2018 год) сочетание жанров Action|Adventure ни разу не получало среднюю оценку ниже 3.
# B Наилучшую оценку жанр Action|Adventure|Animation|Children|Comedy|IMAX получил в 2010 году.
# C Среди сочетаний жанров, получивших наивысшую среднюю оценку в 2018 году, есть сочетание Animation|Children|Mystery.
# D Для жанра Comedy прослеживается тенденция падения рейтинга с каждым годом (с 1996 по 2018).

# Ответ Верно:
# A Если посмотреть оценки жанра Action|Adventure, обратившись к сводной таблице по этому столбцу, оценок ниже 3 не оказывается.
# C Если обратиться к сводной таблице по индексу 2018 года и отсортировать результат по убыванию, то среди жанров, получивших наивысшую оценку, можно найти сочетание Animation|Children|Mystery.
```
